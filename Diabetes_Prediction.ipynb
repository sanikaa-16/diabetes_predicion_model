```python
#import the dependencies
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and analysis

```python
diabetes_data = pd.read_csv('/content/diabetes.csv')
```

```python
diabetes_data.head()
```

```python
diabetes_data.shape
```

```python
#getting statistical measures of the data
diabetes_data.describe()
```

```python
diabetes_data['Outcome'].value_counts()
```

0 ---> Non diabetic
1 ---> diabetic

```python
diabetes_data.groupby('Outcome').mean
```

```python
# seperating the data and labels ---> axis =1 for col and 0 for row

X = diabetes_data.drop(columns = 'Outcome', axis = 1)
Y = diabetes_data['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

Data Standardisation

```python
#standardise the data---> because there is a wide range of numbers it'll be difficult to make a prediction
scalar = StandardScaler()
scalar.fit(X)
```

```python
standard_data = scalar.transform(X)
```

```python
print(standard_data) #we see a range between 0 and 1
```

```python
X = standard_data
Y = diabetes_data['Outcome']
```

Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 2, stratify = Y)
#imp : stratify Y is used to ensure that the outcomes 0/1 are equally distributed amoung train and test model
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Train the Model

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
#training the support vector machine classifier (svm)
classifier.fit(X_train, Y_train)
```

MODEL EVALUATION

accuracy score on train data

```python
#accuracy score on the training data
#this predicts the label for x_train
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy score of the training data")
print(training_data_accuracy)
```

Accuracy score on test data

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy of test data")
print(test_data_accuracy)
```

Making a predictive system

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

input_data = (4,110,92,0,0,37.6,0,191,30) # original input data with 9 features

# Assuming the StandardScaler was trained with 8 features,
# we need to select only 8 features from the input data

# Choose the 8 features you want to use.
# For example, to drop the last feature, use:
input_data_8_features = input_data[:-1]

#change it to a numpy array
input_as_np = np.asarray(input_data_8_features)

#reshape the array as we are predicting for one instance
input_reshaped = input_as_np.reshape(1,-1)

#standardise the input data
std_data = scalar.transform(input_reshaped) # Now the input has 8 features
print(std_data)
```

```python
prediction = classifier.predict(std_data)
print(prediction)
```

```python
if(prediction[0]== 0):
  print("You are not diabetic")
else:
  print("You are diabetic")
```
